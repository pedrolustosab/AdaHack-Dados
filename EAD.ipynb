```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

```python
df = pd.read_csv('base_de_dados.csv', encoding='utf-8')
df
```

```python
# Transformando as colunas em snake_case e em minúsculo
df_columns = df.columns
df_columns = [unidecode(x).replace(' ', '_').lower() for x in df_columns]
df_columns 
```

```python
df_columns = [x.replace('raassa', 'raca').lower() for x in df_columns]
df_columns = [x.replace('endereasso', 'endereco').lower() for x in df_columns]
df_columns = [x.replace('formaassapso', 'formacao').lower() for x in df_columns]
df_columns
```

```python
novos_nomes_colunas = dict(zip(df.columns, df_columns))
df = df.rename(columns=novos_nomes_colunas)
df
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == np.object:  # Se a coluna for textual
        df[col] = df[col].apply(lambda x: unidecode(str(x)).upper())
```

```python
df['genero'] = df['genero'].replace({'FEM': 'F', 'MASC': 'M'})
```

```python
df = df.replace([np.inf, -np.inf, np.nan], -1)
df['idade'] = df['idade'].astype(int)
df['idade'].unique()
```

```python
df.loc[df['tempo_de_casa'] < 0, 'tempo_de_casa'] = -1
df['tempo_de_casa'] = df['tempo_de_casa'].astype(int)
df['tempo_de_casa'].unique()
```

```python
df['formacao'].unique()
```

```python
# Padronizando o logradouro com uso de dicionário
mapeamento_formacao = {
    'ENSINO MA(C)DIO': 'ENSINO MEDIO',
    'ENSINO SUPERIOR': 'ENSINO SUPERIOR',
    'PA3S GRADUAASSAPSO': 'POS GRADUACAO',
    'MESTRADO': 'MESTRADO',
    'DOUTORADO': 'DOUTORADO',
    'NAN': 'NAO INFORMADO'
}

df['formacao'] = df['formacao'].replace(mapeamento_formacao)
df['formacao'].unique()
```

```python
# Dicionário com as correções que queremos fazer
mapeamento_estado = {
    'SANTA CATARINA': 'SC',
    'PARA!': 'PA',
    'TOCANTINS': 'TO',
    'CEARA!': 'CE',
    'SAPSO PAULO': 'SP',
    "RONDA'NIA": 'RO',
    'AMAPA!': 'AP',
    'MINAS GERAIS': 'MG',
    'ACRE': 'AC',
    'RORAIMA': 'RR',
    'AMAZONAS': 'AR',
    'PARANA!': 'PR',
    'BAHIA': 'BA',
    'PIAUA': 'PI',
    'GOIA!S': 'GO',
    'RIO DE JANEIRO': 'RJ',
    'RIO GRANDE DO NORTE': 'RN',
    'PERNAMBUCO': 'PE',
    'MATO GROSSO': 'MT',
    'RIO GRANDE DO SUL': 'RS',
    'ESPARITO SANTO': 'ES',
    'MARANHAPSO': 'MA',
    'PARAABA': 'PB',
    'DISTRITO FEDERAL': 'DF',
    'MATO GROSSO DO SUL': 'MS',
    'SERGIPE': 'SE',
    'ALAGOAS': 'AL'
}

# Substituindo os valores do dicionário no dataframe
df['estado'] = df['estado'].replace(mapeamento_estado)
df['estado'].unique()
```

```python
df['departamento'] = df['departamento'].replace({'OPERAASSAUES': 'OPERACOES'}).str.upper()
df['departamento'].unique()
```

```python
df['senioridade'].unique()
df['senioridade'] = df['senioridade'].replace({'ANALISTA JAONIOR': 'ANALISTA JUNIOR', 'ANALISTA SAANIOR': 'ANALISTA SENIOR','ESTAGIA!RIO': 'ESTAGIARIO' }).str.upper()
df['senioridade'].unique()
```

```python
df = df[['genero', 'idade', 'raca', 'formacao', 'estado', 'tempo_de_casa', 'departamento', 'senioridade' ]]
```
